#### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
df.info()
```

```python
df4 = df[['id', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
          'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']]
df4
```

```python
corr_matrix = df4.corr()
corr_matrix
```

```python
upp_mat = np.triu(df4.corr())
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm', mask= upp_mat);
```

Наиболее коррелируемые признаки, влияющие на стоимость недвижимости это:
1. Жилая площадь (sqft_living)
2. Оценка внешнего вида (grade)
3. Количество ванных комнат (bathrooms)
4. Площадь выше уровня земли (sqft_above)
5. Количество жилой площади у 15-ти соседей (sqft_living15)
6. Количество спален (bedrooms)
7. Вид (view)
8. Количество этажей (floors )
9. Наличие набережной (waterfront)
10. Наличие подвала (sqft_basement)
11. Широта (lat)

```python
sns.lineplot (x=df4['sqft_living'], y=df4['price'])
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.title('Распределение цен в зависимости от жилой площади');
```

Прямая зависимость цены от количества квадратных метров жилой площади, при этом цена резко идет вниз, когда жилая площадь становится больше 12 000 м2

```python
sns.jointplot(x = df4['sqft_living15'], y = df4['price'], kind = 'reg', height=8, ratio=4)
plt.title('Распределение цен в зависимости от жилой площади у 15-ти соседей');
```

Прямая зависимость цены от количества квадратных метров жилой площади у 15-ти соседей

```python
g = sns.JointGrid(data=df4, x="bathrooms", y="price",height=8)
g.plot_joint(sns.scatterplot, s=100, alpha=.5)
g.plot_marginals(sns.histplot, kde=True)
plt.title('Распределение цен в зависимости от количества ванных комнат');
```

Также наблюдается увеличение цены при увеличении количества ванных комнат

```python
plt.figure(figsize=(10, 4))
plt.scatter(df4['sqft_above'], df4['price'])
plt.xlabel('sqft_above')
plt.ylabel('Price')
plt.title('Распределение цен в зависимости от колличества кв.м над уровнем земли');
```

Тут также прямая зависимость

```python
sns.boxplot(x=df4['grade'], y=df4['price'], whis=1.0)
plt.xlabel('Grade')
plt.ylabel('Price')
plt.title('Распределение цен в зависимости от класса дома');
```

С увеличением уровня класса растет и разница в цене. Для домов с 7-11 классом наблюдается максимальное количество выбросов.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df4['price'], y=df4['waterfront'].astype('str'), whis=1.5, hue=df4['grade'])

plt.title('Распределение цен по береговой линии')
plt.xlabel('Price')
plt.ylabel('Waterfront');
```

- уровень цен, не имеющих набережную, значительно ниже, чем цена домов, имеющих ее
- разбег цены для домов с набережной больше, чем для домов без нее.

```python
sns.boxplot(x = df4['price'], y = df4['floors'].astype('str'), whis = 1.5)
plt.xlabel('Price')
plt.ylabel('Floors')
plt.title('Распределение цены в зависимости от этажности');
```

- наибольший разбег по ценам наблюдается для домов в 2,5 этажей;
- минимальный разбег по ценам наблюдается для домов в 3,5 этажей;
- по 2-этажным домам наблюдается максимальное количество выбросов;
- дома с 3,5 этажами почти не имеют выбросов;
- самый дорогой дом имеет 2,5 этажа
